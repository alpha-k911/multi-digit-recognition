```python
def preprocessing(im, border=10):
#     im = cv2.imread('../second/cells/'+i,0)
    _,th1 = cv2.threshold(im, 127,255, cv2.THRESH_OTSU)
#     plt.imshow(255-th1,cmap='gray');
    img = 255-th1
    h,w = img.shape
    mask = np.ones(img.shape[:2], np.uint8)*0
    mt = border
    mask[mt:h-mt, mt:w-mt] = 255
    # display(mask)
    img = cv2.bitwise_and(img,img,mask = mask)
#     display(img)
    return img
```

```python
def preprocess_and_get_counters(img):
    #opencv
    #finding rectangles
#     img = cv2.imread('../second/cells/img967_0_.jpg',0)
#     img = cv2.imread('../second/cells/img967_0_.jpg',0)
    h,w = img.shape
    r = h/w
#     print(h,w,r)
    if h > 72 and w > 72:
        s = h//72
        scale_percent = 100//s 
        width = int(img.shape[1] * scale_percent / 100)
        height = int(img.shape[0] * scale_percent / 100)
        dim = (width, height)
        img = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
        print(width,height,100//s)
    img = preprocessing(img, border=12)
    #upto_masking and otsu
    
    
    canny_output = cv2.Canny(img,1,0)
    display(img)
    display(canny_output)
    contours, hierarchy = cv2.findContours(canny_output, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    # Draw contours
    drawing = np.zeros((canny_output.shape[0], canny_output.shape[1], 3), dtype=np.uint8)
    rec = []
    print("##########")
    for contour in contours:
        (x,y,w,h) = cv2.boundingRect(contour)
        rec.append((x,y,w,h))
#         print(x,y,w,h)
        print("Contour Area:",cv2.contourArea(contour))
        cv2.rectangle(drawing, (x,y), (x+w,y+h), (0,255,0), 2)
    
    print("##########")
    display(drawing)
    
    return contours,img
```

```python
def get_rectangles_from_contours(contours,img):
    rects = []
    for contour in contours:
        (x,y,w,h) = cv2.boundingRect(contour)
        x1=x
        y1=y
        x2=x+w
        y2=y+h
        rects.append((x1,y1,x2,y2))
#     print(rects)
    rects.sort(key=lambda x: x[0])
#     print(rects)
    drawing = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
    for r in rects:
        (x,y,x2,y2) = r
    #     rec.append((x,y,w,h))
#         print(x2-x,y2-y)
        cv2.rectangle(drawing, (x,y), (x2,y2), (0,255,0), 2)
        display(drawing)
    return rects
```

```python
def group_together(rects,img):
    p = 0
    grouped = []
    l = len(rects)
    i  = 0
#     print(l)
    while i != len(rects)-1:
        
        (x1,y1,x2,y2) = rects[i]
#         print("checking for:",rects[i], ":", i)
        (p1,q1,p2,q2) = rects[i+1]
#         if x1 <= p1 <= x2 and y1 <= q1 <= y2:
        if x1 <= p1 <= x2:# and y1 <= q1 <= y2:
#             print('yes')
            xx1 = min(x1,p1)
            yy1 = min(y1,q1)
            xx2 = max(x2,p2)
            yy2 = max(y2,q2)
#             print("previous:",rects[i])
            rects[i] = (xx1,yy1,xx2,yy2)
#             print("new:",rects[i])
#             print("deleting:",rects[i+1])
            rects.pop(i+1)
        else:    
            i = i+1
#         print("O ")
    display(img)
    drawing = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
    for r in rects:
        (x,y,x2,y2) = r
    #     rec.append((x,y,w,h))
#         print(x2-x,y2-y)
        cv2.rectangle(drawing, (x,y), (x2,y2), (0,255,0), 2)
        display(drawing)
    return rects
# group_together(rects)
```

```python
def pad_to_square(rects,img,thresh=10):
    imgs = []
    h_, w_ = 32,32
    
    a = 0
    for i in rects:
        (x,y,x2,y2) = i
        
        area = (y2-y)*(x2-x)
        print("Area:", area)
        a+= area
    print("Total area:", a, a/len(rects))
    thresh = (a/len(rects))/12
    thresh = 10
    
    for i in rects:
        h_, w_ = 32,32
        (x,y,x2,y2) = i
        
        area = (y2-y)*(x2-x)
        if area >= thresh:
            while y2-y > h_ or x2-x > w_:
                h_ += 16
                w_ += 16
    #             print("clash:", y2-y,x2-x, h_)
    #             print("changing dims to:", h_)
            zero = np.zeros((h_,w_))
            ver_pad = w_ - (y2-y)
            hor_pad = h_ - (x2-x)
            top_pad = ver_pad // 2
            bot_pad = ver_pad - top_pad
            left_pad = hor_pad // 2
            right_pad = hor_pad - left_pad
    #         print(x2-x,y2-y)
    #         print(w_ - right_pad - left_pad,h_-bot_pad - top_pad)
            display(img[y:y2,x:x2])
            zero[top_pad:h_- bot_pad, left_pad:w_ - right_pad] = img[y:y2,x:x2]
            display(zero)
    #         print(zero.shape)
            zero = cv2.resize(zero, (28,28))
            display(zero)
            imgs.append(zero)
#             print("-----------------")
#             print("Rectangle Area:",((y2-y)*(x2-x)))
#             print("-----------------")
#         print("Done with:", i)
    return imgs
# display(img)
# imgs = pad_to_square(rects)
```

# All-Together

```python
def predict_cell(img,label,miss_list):
    #opencv
    #finding rectangles
    
#     blockPrint()
    
    counter = 0
    con, img = preprocess_and_get_counters(img)

    #sorting rectangles
    rectangles = get_rectangles_from_contours(con,img)

    #merging rectangles
    rectangles = group_together(rectangles,img)

    #padding to make square of size 28x28
    imgs = pad_to_square(rectangles,img)

    #convertion
    imgs_list = imgs
    imgs = np.array(imgs)
    imgs = np.reshape(imgs,(imgs.shape[0] ,28,28,1))
    # print(imgs[0].shape)

    #prediction
    preds = model.predict(imgs)
    
#     enablePrint()

    #seeing preds
    prediction = ''
    for i in range(len(preds)): 
        m = preds[i].argmax()
#         print(m)
        if m != 10:
            prediction += str(m)
        else:
            prediction += '.'
        display_(imgs_list[i], str(preds[i].argmax()))
    
    print(">>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>")
    
    p = prediction
    try:
        while prediction[-1] == '.':
            prediction = prediction[:-1]
        while prediction[0] == '.':
            prediction = prediction[1:]
        p = int(prediction)
        label = int(label)
    except:
        p = prediction
        label = str(label)
    
    if p == label:
        counter += 1
        print("Correct Predicton for: ", "[",label, "]")
    else:
        print("Failed Predicton for: ", label, "[",prediction, "]")
        miss_list.append(str(label)+" : "+str(prediction))
        
    print(">>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>")
        
    return counter
```

```python
pp = 0
miss_list = []
path = "../second/cells/"
f = os.listdir(path)
total = len(f)
hits = 0
ex_time = []
tic_ = timeit.default_timer()
for j in range(186):
#     i = "img157_1.5_.jpg"
    i = f[j]
    print("Checking:",i)
    tic = timeit.default_timer()
    im = cv2.imread(path+i,0)
    hits += predict_cell(im, i.split('_')[1], miss_list)
    toc = timeit.default_timer()
    ex_time.append(toc-tic)
toc_ = timeit.default_timer()
ex_time = np.array(ex_time)
pp = 1
print("")
print("----------------------------- RESULTS ----------------------------------")
print("")
print("Accuracy:", hits/total)
print("Total no. of images tested:", total)
print("Min execution time:", ex_time.min())
print("Avg execution time:", (toc_-tic_)/total)
print("Max execution time:", ex_time.max())
print("Total execution time:", toc_-tic_)
```
